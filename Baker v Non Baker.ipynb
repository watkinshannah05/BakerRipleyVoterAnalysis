```python
def mse(result, expected):
    """
    Calculate the mean squared error between two data sets.

    The length of the inputs, result and expected, must be the same.

    inputs:
        - result: a list of integers or floats representing the actual output
        - expected: a list of integers or floats representing the predicted output

    returns: a float that is the mean squared error between the two data sets
    """
    # will hold the sum of the squared errors
    total = 0
    # get the squared errors and add them to total for each individual prediction
    for index in range(len(result)):
        diff = result[index] - expected[index]
        total += (diff ** 2)
    # get the mean squared error
    return total/len(result)
```

```python
mse_data = {}
for precinct in ripley_nums:
    mse_data[precinct] = []


x = 0
while x < non_ripley_precincts.shape[0]:
    row = non_ripley_precincts.iloc[x]
    result = list(row[9:])
    y = 0
    while y < ripley_precincts.shape[0]:
        rip = ripley_precincts.iloc[y]
        expected = list(rip[9:])
        diff = mse(result, expected)
        mse_data[int(rip[0])].append(diff)
        y += 1
    x += 1
    
```

```python
# All remaining precincts are comparable to the target precincts. So, 15 will be selected randomly to compare to the original 15 targeted by BakerRipley.
```

```python
# Find the most similar precinct
similar = []
list()
for key in mse_data:
    
    dat = mse_data[key]
    ind = dat.index(min(dat))
    precinct = non_ripley_precincts.iloc[ind][0]
    similar.append(precinct)

    
print(ripley_nums)
print(similar)
```

```python
similar_precincts = non_ripley_precincts[non_ripley_precincts.Precincts.isin(similar) == True]
similar_precincts
```

```python
voters_1 = pd.read_csv('nov_voters_1.csv', low_memory=False)
voters_2 = pd.read_csv('nov_voters_2.csv', low_memory=False)
voters_3 = pd.read_csv('nov_voters_3.csv', low_memory=False)

temp = pd.concat([voters_1, voters_2])
voters = pd.concat([temp, voters_3])
voters.drop(['Unnamed: 0', 'DATE_VOTED', 'Voting Site', 'Congressional House District'], inplace=True, axis=1)

voters.head()
```

```python
rip_rates = []
sim_rates = []
for idx in range(len(similar)):
    rip = voters[voters['Voting Precinct'] == ripley_nums[idx]].shape[0]
    sim = voters[voters['Voting Precinct'] == similar[idx]].shape[0]
    rip_vap = int(ripley_precincts.iloc[idx][2].replace(',', ''))
    rip_rate = round((rip/rip_vap)*100, 2)
    temp = similar_precincts[similar_precincts['Precincts'] == similar[idx]]
    sim_vap = int(temp.iloc[0][2].replace(',', ''))
    sim_rate = round((sim/sim_vap)*100, 2)
    rip_rates.append(rip_rate)
    sim_rates.append(sim_rate) 
```

```python
labels = [("#"+str(rip), "#"+str(sim)) for rip, sim in zip(ripley_nums, similar)]
print(labels)

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
rip_bars = ax.bar(x - width/2, rip_rates, width, label="Baker Ripley Targeted Precincts", color='#3d54c8')
sim_bars = ax.bar(x + width/2, sim_rates, width, label="Other Harris County Precincts", color='#95e1e9')

ax.set_ylabel('Voter Turnout')
ax.set_title('Voter Turnout in Baker Ripley Targeted Precincts vs Other Harris County Precincts')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rip_bars, padding=1.5)
ax.bar_label(sim_bars, padding=1.5)
ax.set_ylim(0, 100)

for ax in fig.axes:
    ax.yaxis.set_major_locator(MultipleLocator(5))
    
fig.set_size_inches(20, 10)
fig.tight_layout()
plt.show()

```
