```python
total_pop = target_precincts[target_precincts['Population'] == 'Total:']
va_pop = target_precincts[target_precincts['Population'] == 'VAP:']
total_pop = total_pop.reset_index()
total_pop.drop(['index'], axis=1, inplace=True)
total_pop
```

```python
total_pop.drop(['Population'], axis=1, inplace=True)
va_pop.drop(['Population'], axis=1, inplace=True)
total_pop.rename(columns = {'Total':'Total Population'}, inplace = True)
total_pop['VAP'] = va_pop['Total']
total_pop
```

```python
total_pop = total_pop.apply(pd.to_numeric, errors='ignore')
total_pop.dtypes
```

```python
total_pop['Total Population']=total_pop['Total Population'].str.replace(',','')
total_pop
total_pop['Total Population'] = pd.to_numeric(total_pop['Total Population'])
total_pop
```

```python
total_pop['VAP']=total_pop['VAP'].str.replace(',','')
total_pop['VAP'] = pd.to_numeric(total_pop['VAP'])
total_pop
```

```python
sns.set_style("darkgrid")
sns.set_context('talk')
plt.figure(figsize=(15,10))
sns.scatterplot(x="Precincts",
               y='VAP',
               size='Total Population',
               data=total_pop,
                hue="%Non-Anglo",
                sizes=(50,2000),
                palette=("GnBu_d"))

plt.xlabel("Precinct Number")
plt.ylabel("Voter Age Population (VAP)")
plt.legend(bbox_to_anchor=(1, 1), loc='upper left', fontsize=17)
plt.tight_layout()
plt.show()
```
